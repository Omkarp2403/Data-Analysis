```python
df.rename(columns = {'IncidentID*+' : 'Incident_ID',
                    'Submitter*' : 'Submitter',
                    'SubmitDate' : 'Submit_Date',
                    'Summary*' : 'Summary',
                    'Service*+' : 'Service',
                    'Priority*' : 'Priority',
                    'Status*' : 'Status',
                    'SLMRealTimeStatus':'SLM_Real_Time_Status',
                    'Assignee+' : 'Assignee',
                    'Group Owner' : 'Group_Owner',
                    'AssignedGroup*+' : 'Assigned_Group',
                    'Detailed Description' : 'Detailed_Description',
                    'LastModifiedBy' : 'Last_Modified_By',
                    'AssignmentLastModifiedDate' : 'Assignment_Last_Modified_Date',
                    'LastResolvedDate' : 'Last_Resolved_Date',
                    'FirstName' : 'First_Name',
                    'GroupTransfers' : 'Group_Transfers',
                    'Resolution' : 'Resolution',
                    'LastName' : 'Last_Name',
                    'ReportedSource' : 'Reported_Source'},
                    inplace = True)
```
